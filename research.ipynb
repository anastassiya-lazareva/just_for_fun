```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Датасет.csv')
df = df.set_index('Наименование')
df.columns = pd.to_datetime(df.columns, dayfirst=True)
df.head()
```

```python
#создадим датасет с темпами прироста цен 2022, 2023, 2024, 2022-2025
df_dropna = df.dropna()

columns_2022 = [col for col in df_dropna.columns if col.year == 2022]
columns_2023 = [col for col in df_dropna.columns if col.year == 2023]
columns_2024 = [col for col in df_dropna.columns if col.year == 2024]
columns_2025 = [col for col in df_dropna.columns if col.year == 2025]

growth_rate = pd.DataFrame()

growth_rate['2022'] = (df_dropna[max(columns_2022)] - df_dropna[min(columns_2022)])/ df_dropna[min(columns_2022)]*100
growth_rate['2023'] = (df_dropna[max(columns_2023)] - df_dropna[min(columns_2023)])/ df_dropna[min(columns_2023)]*100
growth_rate['2024'] = (df_dropna[max(columns_2024)] - df_dropna[min(columns_2024)])/ df_dropna[min(columns_2024)]*100
growth_rate['2022-2025'] = (df_dropna[max(columns_2025)] - df_dropna[min(columns_2022)])/ df_dropna[min(columns_2022)]*100

growth_rate.head()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=True)

years = ['2022', '2023', '2024', '2022-2025']
colors = ['skyblue', 'lightgreen', 'salmon', 'orchid']

axes = axes.flatten()

all_data = pd.concat([growth_rate[year] for year in years])
x_min, x_max = all_data.min(), all_data.max()

for i, year in enumerate(years):
    sns.histplot(growth_rate[year], kde=True, bins=20, color=colors[i], ax=axes[i])

    mean_value = growth_rate[year].mean()

    axes[i].axvline(mean_value, color='black', linestyle='--', linewidth=2, label=f'Среднее: {mean_value:.2f}%')
    axes[i].legend() 

    axes[i].set_xlim(x_min, x_max)
    axes[i].set_title(f'Прирост цен ({year})', fontsize=12)
    axes[i].set_xlabel('Прирост (%)')

    if i % 2 == 0:
        axes[i].set_ylabel('Кол-во товаров')
    else:
        axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```
